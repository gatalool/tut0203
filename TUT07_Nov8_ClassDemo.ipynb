## <span style="color:#1E3765">1.B. A simple linear regression</span>

```python
import pandas as pd
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt

# Load the data from the CSV file
df = pd.read_csv('school_data_250.csv')

# Perform a simple linear regression using smf
# Let's predict MathScore based on EnglishScore
model = smf.ols('MathScore ~ EnglishScore', data=df).fit()

# Print the summary of the regression model
print(model.summary())

# Generate a scatter plot with the regression line
plt.figure(figsize=(10, 6))
plt.scatter(df['EnglishScore'], df['MathScore'], color='blue', label='Data points')
plt.plot(df['EnglishScore'], model.fittedvalues, color='red', label='OLS Regression Line')
plt.xlabel('English Score')
plt.ylabel('Math Score')
plt.title('OLS Regression: Math Score vs. English Score')
plt.legend()
plt.grid(True)
plt.show()
```

## What are the variables in this question?
- what is the exposure/independent variable?

<span style="font-family:Times MS"> English Score </span>

- what is the outcome?

<span style="font-family:Times MS"> Math Score </span>


## What is the hypothesis of this question?

<span style="font-family:Times MS"> Ho: There is no significant association with and alpha of 0.05 between English Score and Math Score Ha: There is a significant relationship between English Score and Math Score </span>

### what are the key results of this question?

<span style="font-family:Times MS"> There is no significant relationship between Math Score and English Score (We cannot reject H0)

*suppose there was a significant relationship: For every 1 mark increase in English Score there is -0.0624 mark decline Math Score </span>

## <span style="color:#1E3765">1.C.a. A multiple linear regression using both the continuous and binary variables</span>

```python
import seaborn as sns
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Perform a multiple linear regression using smf
# Let's predict MathScore based on EnglishScore, ScienceScore, and FitnessTest
model = smf.ols('MathScore ~ EnglishScore + ScienceScore + FitnessTest', data=df).fit()

# Print the summary of the regression model
print(model.summary())

# Generate a pairplot to visualize the relationships between variables
sns.pairplot(df, vars=['MathScore', 'EnglishScore', 'ScienceScore'], hue='FitnessTest')
plt.suptitle('Pairplot of Math Score, English Score, and Science Score by Fitness Test', y=1.02)
plt.show()

# Generate a scatter plot with the regression plane (3D plot)
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot
ax.scatter(df['EnglishScore'], df['ScienceScore'], df['MathScore'], c='blue', marker='o')

# Create a meshgrid for the regression plane
xx, yy = np.meshgrid(np.linspace(df['EnglishScore'].min(), df['EnglishScore'].max(), 100),
                     np.linspace(df['ScienceScore'].min(), df['ScienceScore'].max(), 100))
zz = model.params[0] + model.params[1] * xx + model.params[2] * yy + model.params[3] * (df['FitnessTest'] == 'Pass').mean()

# Plot the regression plane
ax.plot_surface(xx, yy, zz, color='red', alpha=0.5)

ax.set_xlabel('English Score')
ax.set_ylabel('Science Score')
ax.set_zlabel('Math Score')
ax.set_title('Multiple Linear Regression: Math Score vs. English Score and Science Score')

plt.show()
```

## What are the variables in this question?
- what is the exposure/indepenent variable?

<span style="font-family:Times MS">Math Score</span>

- what is the outcome/dependent variable?

<span style="font-family:Times MS">English Score</span>

- what are the additional variables (covariates)?

<span style="font-family:Times MS">Science Score and FitnessTest</span>

## What is the hypothesis of this question?

<span style="font-family:Times MS">Ho: There is no significant association with and alpha of 0.05 between Math Score and English Score, while adusting for Science Score and Fitness Test 
    
<span style="font-family:Times MS">Ha: There is a significant relationship between English Score and Math Score, while adusting for Science Score and Fitness Test </span>

### what are the key results of this question?

<span style="font-family:Times MS"> There is no significant association between Math Score and English Score when adjusting for Fitness Test and Science Score
    
<span style="font-family:Times MS"> Let's assume there was a significant relationship: For every 1 mark increase English Score, there is a 0.0056 mark increase in Math Score, while adjusting for Science SCore and Fitness Test</span>

**Asking Copilot:** *Can you produce a linear equation when the exposure is a continuous variable there is a binary interaction term and there is a second continuous independent variable?*

Sure! Let's construct a linear regression model with the following components:

1. **Exposure (X1)**: A continuous variable.
2. **Binary Interaction Term (X2)**: A binary variable (0 or 1).
3. **Second Continuous Independent Variable (X3)**: Another continuous variable.

The general form of the linear regression equation can be written as:

$$ Y = \beta_0 + \beta_1 X1 + \beta_2 X2 + \beta_3 X3 + \beta_4 (X1 \times X2) + \epsilon $$

Where:
- \( Y \) is the dependent variable.
- \( \beta_0 \) is the intercept.
- \( \beta_1, \beta_2, \beta_3, \beta_4 \) are the coefficients for the respective variables.
- \( \epsilon \) is the error term.
- \( X1 \times X2 \) represents the interaction term between the continuous variable \( X1 \) and the binary variable \( X2 \).

This equation allows you to assess the main effects of \( X1 \) and \( X3 \), the effect of the binary variable \( X2 \), and the interaction effect between \( X1 \) and \( X2 \).
